```python
import pandas as pd
df=pd.read_csv("/kaggle/input/hourly-energy-consumption/PJMW_hourly.csv")
df=df.set_index('Datetime')
df.index=pd.to_datetime(df.index)

df.describe()
```

```python
#Defining required fields
df['year']=[x for x in df.index.year]
df['month']=[x for x in df.index.month]

df=df.reset_index() #change the index into a column and default index of number is set

#apply methods used to apply a function on pandas axis of dataframe or series
df['week']=df['Datetime'].apply(lambda x:x.week)
df=df.set_index('Datetime')
df['hour']=[x for x in df.index.hour]
df['day']=[x for x in df.index.day_of_week]
df['day_str']=[x.strftime('%a') for x in df.index] #Formats a datetime object as a string representing the abbreviated weekday name.
df['year_month']=[str(x.year)+'_'+str(x.month) for x in df.index]
```

### Time Series Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose

df_plot = df[df['year'] == 2017].reset_index()
df_plot = df_plot.drop_duplicates(subset=['Datetime']).sort_values(by='Datetime')
df_plot = df_plot.set_index('Datetime')
df_plot['PJME_MW_Multiplicative_Decompose'] = df_plot['PJMW_MW']
df_plot['PJME_MW_Additive_Decompose'] = df_plot['PJMW_MW']

# Additive Decomposition
result_add = seasonal_decompose(df_plot['PJME_MW_Additive_Decompose'], model='additive', period=24*7)

# Multiplicative Decomposition 
result_mul = seasonal_decompose(df_plot['PJME_MW_Multiplicative_Decompose'], model='multiplicative', period=24*7)

# Plot
result_add.plot().suptitle('', fontsize=22)
plt.xticks(rotation=45)
result_mul.plot().suptitle('', fontsize=22)
plt.xticks(rotation=45)
plt.show()
```

The above plots refers to 2017. In both cases, we see the trend has several local peaks, with higher values in summer. From the seasonal component, we can see the series actually has several periodicities, this plot highlights more the weekly one, but if we focus on a particular month (January) of the same year, daily seasonality emerges too

```python
df_plot = df[(df['year'] == 2017)].reset_index()
df_plot = df_plot[df_plot['month'] == 1]
df_plot['PJMW_MW_Multiplicative_Decompose'] = df_plot['PJMW_MW']
df_plot['PJMW_MW_Additive_Decompose'] = df_plot['PJMW_MW']

df_plot = df_plot.drop_duplicates(subset=['Datetime']).sort_values(by='Datetime')
df_plot = df_plot.set_index('Datetime')

# Additive Decomposition
result_add = seasonal_decompose(df_plot['PJMW_MW_Additive_Decompose'], model='additive', period=24*7)

# Multiplicative Decomposition 
result_mul = seasonal_decompose(df_plot['PJMW_MW_Multiplicative_Decompose'], model='multiplicative', period=24*7)

# Plot
result_add.plot().suptitle('', fontsize=22)
plt.xticks(rotation=45)
result_mul.plot().suptitle('', fontsize=22)
plt.xticks(rotation=45)
plt.show()
```

### Lag Analysis
1. Lag Analysis:
     * `Lag`: Refers to past values of a time series. For instance, in a daily series, a lag of 1 represents the value from the previous day, a lag of 2 represents the value from two days ago, and so forth.
     *  `Autocorrelation`: Measures how a time series is correlated with a lagged version of itself. For example, autocorrelation at lag 1 tells you how today's value is related to yesterday's value.
2. Autocorrelation Function (ACF):
     * `ACF`: A plot that shows autocorrelation coefficients for various lags. If the plot shows high values for specific lags, it indicates that the time series has a pattern or seasonality at those lags.
3. Partial Autocorrelation Function (PACF):
     * `PACF`: Focuses on the direct correlation between a time series and its lags, removing the effects of intermediate lags. For example, PACF at lag 3 measures the correlation between the series and its value three periods ago, excluding the influence of lags 1 and 2.

```python
from statsmodels.graphics.tsaplots import plot_pacf

actual = df['PJMW_MW']
hours = range(0, 24, 4)

for hour in hours:
    plot_pacf(actual[actual.index.hour == hour].diff().dropna(), lags=30, alpha=0.01)
    plt.title(f'PACF - h = {hour}')
    plt.ylabel('Correlation')
    plt.xlabel('Lags')
    plt.show()
```

### Conclusion:
he PACF simply consists on plotting Pearson partial auto-correlation coefficients for different lags. Of course, the non-lagged series shows a perfect auto-correlation with itself, so lag 0 will always be 1. The blue band represent the confidence interval: if a lag exceed that band, then it is statistically significant and we can assert it is has great importance.

### Feature Engineering
* Lag Analysis is crucial for improving time series models. By examining how past values (lags) relate to current values, we can determine which lags are important for forecasting.

* Hourly Division: Splitting data into 24 subsets, each representing an hour of the day, helps to regularize and smooth the data. This makes forecasting simpler and more accurate.

* Model Training: Each hourly subset should be individually analyzed, trained, and fine-tuned. The final forecast is obtained by combining predictions from these 24 models. Each hour's model will focus on relevant lags.

#### Types of Lags:

* Auto-regressive Lags: These are lags close to the present. They show how recent values affect the current value. High values for these lags indicate strong trends in the series.
* Seasonal Lags: These lags represent recurring patterns, like weekly cycles. When splitting data by hour, these often reflect weekly seasonality.

#### Night Hours
Consumption during night hours (0, 4) relies more on recent values (auto-regressive lags) rather than seasonal patterns. For these hours, focus on the first five lags since seasonal lags (like 7, 14 days) are less significant.

#### Day Hours
During day hours (8, 12, 16, 20), both recent and seasonal lags are important. Hours like 8 and 12 show high consumption and strong seasonal patterns. For these hours, include both types of lags to capture trends and seasonal effects accurately.

### Tips:
* Avoid Overfitting: Dont use too many lags in your model, as this can lead to overfitting, where the model performs well on training data but poorly on new data. Typically, use auto-regressive lags from 1 to 7 days and weekly lags like 7, 14, 21, and 28 days. However, its not necessary to include all of them.

* Relevance of Lags: Using lags that are neither auto-regressive nor seasonal can cause overfitting. Focus on understanding why a lag is important before including it in the model.

* Transforming Lags: Combining lags into single features can enhance the model. For instance, seasonal lags can be averaged using a weighted mean to summarize the seasonality of the series in a more effective way.
