## Important Details
* What is RandomForestClassifier?
* Ans: **Random Forest Classifier** is a popular ensemble learning algorithm used for classification tasks.Its based on the concept of decision tree,where multiple decision tree are trained on different sub-samples of the datasets and thier predictions are combined to make the final prediction.
 1. **BootStrap Sampling(bagging)**:Random Forest starts by creating multiple decision trees each trained on a random subset of the training data.This process is called is bootstrap sampling.
 2. **Feature Randomness**:In Addition to training each tree on a random subset of the data.Random Forest also selects a random subset of features to consider when splitting each node of the tree.This introduces more randomness into the model and helps to decorrelate the trees.
 3. **Voting**:Once all the trees are trained predictions are made by each individual tree,For classsification tasks,each tree votes for its class label it predicts .The class label with the most votes is chosen as the final prediction.
 * **Effective** because:
   1. **Robustness**:Random Forest is less prone to overfitting compared to individual decision trees because it combines the predictions of multiple trees,each trained on a different subset of the data.
   2. **Handling Imbalanced Data**: Random Forest is a measure of feature importance which can help in understanding the relative importance of different features in making predictions

## Column Transformer
* It is powerfull tool in sklearn used for applying different transformations to different columns or subsets of columns in a dataset.It allows you to specify different preprocessing steps for different columns ,which is particulary useful when dealing with datasets that contain a mixture of numerical and categorical features.
* **Specification of transformer**: You define a set of transformer,each of while specifies the preprocessing steps to be applied to a particular subset of columns.These transformer can be instance of sklearn transformer such as StandardScaler,OneHotEncoder and SimpleImputer etc or custom transformers that you have implemented.
* **ColumnSelection**: For each transformer you specify the columns or column indices to which it should be applied.This allows to apply to different transformations to different subsets of columns in your datasets.
* **Pipeline Construction:** You construct a Column Transformer object by passing a list of tuples where each tuple consists of name for the transformer and transformer iteself along with the columns or column indices to which it should be applied.
* **Fit and Transform**: Once you have defined column transformer you can fit it to your dataset using the 'fit_transform' method.This applies the specifies transformations to the specified columns and returns the transformed dataset.

## Pipeline
### What is Pipeline?
* Ans: A Pipeline is sequence of data processing components or steps that are chained together in a specific order.Each step in the pipeline typically performs some kind of transformation or processing on the data,such as data cleaning feature selection or model training.The output of the one step becomes input of 2 step.
### Why do we need Pipeline?
* Ans: 1.**Organisation**: Pipeline helps organise and structure the workflow of a machine learning project breaking it down into smaller modular components.Each components in the pipeline focues on a specific task, making it understand and manage the overall process.
 2. **Automation**: Pipeline automates the process of applying multiple transformations to the data sequentially .Instead manually applying each transformation one by one ,you can define a pipeline that encapsulates all the preprocessing steps and then apply it to the data in one go.
 3. **Reproducibility**:Pipeline ensure that the same sequence of preprocessing steps is applied consistently to different datasets or subsets of data.This improve the reproducibility of your machine learning experiments and makes it easier to compare different models or datasets.
 4. **Scalability**:Pipeline can be easily extended or modified to incorporate new preprocessing steps or models as needed.This makes it straightforward to experiment with different combinations of preprocessing techniques and alogorithms to find the best approach for your particular problem.

### StandardScaler Vs MinMaxScaler
* **StandardScaler**: It is a technique used in ml to normalise or scale the features of a dataset.
* Mean and StandardDeviation:It Scales the features so that they have a mean of 0 and a standard deviation of 1.
* **Centering Data**: StandardScaler centers the data around 0 by subtracting the mean from each value.
* **Scaling value**:It then scales the centered data by dividing each feature by it std.
* **Preserve Relationships**: StandardScaler preserves the relationships between the features ensuring that the relative differences between the features values are maintained.
* **Effect of Outlier**: It is sensitive to outliers because it use mean and std of the data. Outlier can have a significant impact on the scaling process.
* **Appliability**: StandardScaler is commonly applied to numerical features in the dataset to ensure that all features have a similar scale which can improve the performance of certain machine learning algorithms such as those distance calculations or gradient descent optimization.

### Simple Imputer
* **Handling Missing Data**: Simple Imputer is a tool in sklearn that helps handle missing data in your dataset.
* **Replacement Strategy**: It works by replacing missing values in the dataset with a specified value,such as a mean,median,most frequent value or a constant.
* **Ease of Use**:It is easy to use and can be applied to both numerical and categorical data.
* **Applicability**: SimplieImputer is often used as a preprocessing step in ml pipelines to ensure that the datasets does not contains any missing values before training a model.
* **Default Strategy**:If no streatgy is specified ,SimpleImputer defaults to using the mean straetgy for numerical data and the most frequent strategy for categorical data.
* **Uniform Replacement**: SimpleImputer replaces all missing valuess in the dataset with the same value which is determined based on th specified strategy.

```python
# Define a pipeline for numerical features
numerical_transformer = Pipeline(steps=[
    ("Imputer", SimpleImputer(strategy="mean")),  # Fill missing values with the mean
    ("scaler", MinMaxScaler())                   # Scale numerical features using Min-Max scaler
])

# Define a pipeline for categorical features
categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),  # Fill missing values with the most frequent value
    ("onehot", OneHotEncoder(handle_unknown="ignore"))     # One-hot encode categorical features
])

# Combine numerical and categorical transformers using ColumnTransformer
preprocessor = ColumnTransformer(transformers=[
    ("num", numerical_transformer, new_numerical_cols),    # Apply numerical transformer to numerical features
    ("cat", categorical_transformer, categorical_cols)    # Apply categorical transformer to categorical features
])
```

### 5.K-Nearest Neighbour Classifier
* N-neighbours:Imagine you're trying to predict something for a new data point.This parameter decides how many nearby data points we should look at to make that predictions.If you pick more neighbours ,you get a more stable prediction ,but it might take a longer to compute.
* Distance metric:This is like deciding how you measure the distance between data points.Think of it as choosing between different ways to calculate how 'far' or 'close' two points should have been same say in the predictions or if closer points should have more influence.
* Algorithm:This is about how the computer finds those nearby points.Its like picking the best method for searching through a big book for information you can choose the way that's fastest for your data.
* Leaf Size:Imagine organising data like a tree.This parameters controls how many data points are in each branch of the tree.it can affects how quickly the computers can find the right neighbors.
* P:This is used when we're calculating distances between points.Its like saying how much weight we give to different directions when measuring distance.For example if you're using a map do you care about north-south distance or west east distance?This paramters lets u decide

```python
from sklearn.neighbors import KNeighborsClassifier as KNC

## Define KNeighborsClassifier with specified hyperparameters
model_5 = KNC(
    n_neighbors=20,     # Number of neighbors to use for classification
    weights='distance', # Weight function used in prediction (inversely proportional to distance)
    algorithm='brute'   # Algorithm used to compute the nearest neighbors ('brute' for brute-force search)
)
#
## Define pipeline with preprocessor and KNeighborsClassifier
my_pipeline_5 = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", model_5)
])
#
## Train the pipeline on training data
my_pipeline_5.fit(X_train, y_train)
#
## Make predictions on validation data
preds_5 = my_pipeline_5.predict(X_valid)
#
## Calculate evaluation metrics
confu_5 = confusion_matrix(y_valid, preds_5)
accu_5 = accuracy_score(y_valid, preds_5)
preci_5 = precision_score(y_valid, preds_5, average='micro')
recall_5 = recall_score(y_valid, preds_5, average='micro')
#
## Print evaluation metrics
print("Accuracy 5:", accu_5)
print("Precision 5:", preci_5)
```

### 7.Catboost Classifier
* **Catboost** is a ml library.
* Its designed for boosting , a technique that combines many simple models to make better predictions.
* Its specifically good with data containing both numerical and categorical features.
* CatBoost handles categorical features automatically without needing preprocessing .
* It builds decision trees each correcting the mistakes of the previous ones.
* The combined predictions from all trees usually result in more accurate predictions.
* CatBoost automates many aspects of tree building for better performance.
* **Conclusions**: CAtboost helps build powerful and accurate predictive models with ease.

### Hyperparameters Tuning
* **iterations** - Number of boosting iterations to be run
* **learning_rate** - Step size at each iteration
* **depth** - Maximum depth of each tree
* **l2_leaf_reg** - Strength of L2 Regularisation
* **rsm** -  Random selection of features ratio of each tree
* **loss_function** - Type of loss function to be optimized 
* **border_count** - Number of splits for numerical features 
* **od_wait** - Number of iterations with no significant improvement to stop training for early stopping
* **nan_mode** - How missing values are treated?
* **thread_count** - Number of threads to use for training
* **random_seed** - seed for random number generation 
* **classes_count** - Number of classes in the target variable
* **boosting_type** - Type of Boosting algorithm to use
* **subsample** - Fraction of training data to be used for each tree
* **max_depth** - Maximum depth of trees (alternative to depth)
* **n_estimators** - Number of trees to be build (alternatives to 'iterations')
* **early_stopping_rounds** - Number of rounds with no improvement to trigger earlt stopping 
* **cat_features** - Indices of categorical features
* **task_type** - type of task e.g. 'CPU' or 'GPU'
* **grow_policy** - Tree growing policy e.g. 'SymmetricTree' or 'DepthWise'
* **min_data_in_leaf** - Minimum number of samples in a leaf
* **num_leaves** - Maximum number of leaves in each tree
* **score_function** - Evaluation metric to optimize
* **text_features,tokenizers,dictionaries** - Setting for handling text features
